## 5. Theoretical Implementation

Here is how to implement the Computation Graph logic in Python.


```python
def forward_propagation_graph(a, b, c):
    """
    Computes J = 3(a + bc) using a step-by-step forward pass.
    
    Arguments:
    a, b, c -- Scalar input values
    
    Returns:
    J -- Final output
    u, v -- Intermediate values (needed later for backprop)
    """
    
    # NODE 1: u = bc
    u = b * c
    print(f"Step 1 (u): {b} * {c} = {u}")
    
    # NODE 2: v = a + u
    v = a + u
    print(f"Step 2 (v): {a} + {u} = {v}")
    
    # NODE 3: J = 3v
    J = 3 * v
    print(f"Step 3 (J): 3 * {v} = {J}")
    
    return J, u, v

# --- VERIFICATION ---
if __name__ == "__main__":
    a, b, c = 5, 3, 2
    print(f"Inputs: a={a}, b={b}, c={c}")
    
    J, u, v = forward_propagation_graph(a, b, c)
    
    print(f"\nFinal Result J: {J}")
    # Check: 3 * (5 + (3*2)) = 3 * (5 + 6) = 3 * 11 = 33. Correct.
```
