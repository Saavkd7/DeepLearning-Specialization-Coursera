## Theoretical Implementation (Markdown)



Here is the mathematical logic implemented in NumPy, exactly as described in the theory.


```python
import numpy as np

def sigmoid(z):
    """
    Compute the sigmoid of z

    Arguments:
    z -- A scalar or numpy array of any size.

    Return:
    s -- sigmoid(z)
    """
    # THEORY: Formula 1 / (1 + e^-z)
    # np.exp handles the exponentiation element-wise
    s = 1 / (1 + np.exp(-z))
    return s

def propagate_theory(w, b, X):
    """
    Implement the cost function and its gradient for the propagation step

    Arguments:
    w -- weights, a numpy array of size (n_x, 1)
    b -- bias, a scalar
    X -- data of size (n_x, m) - Note the Column-Major shape!

    Returns:
    A -- The activation (probability vector)
    """
    
    # 1. LINEAR STEP (Z)
    # Matrix Multiplication of w transposed and X
    # Shape: (1, n_x) dot (n_x, m) -> (1, m)
    # Broadcasting adds scalar b to every column
    Z = np.dot(w.T, X) + b
    
    # 2. ACTIVATION STEP (A)
    # Apply sigmoid to squash Z into [0, 1] range
    A = sigmoid(Z)
    
    return A

# --- VERIFICATION OF THEORY ---
if __name__ == "__main__":
    # Example from Transcript: Large Positive vs Large Negative Z
    z_large = 100
    z_small = -100
    z_zero = 0
    
    print(f"Sigmoid(100)  should be ~1.0: {sigmoid(z_large)}")
    print(f"Sigmoid(-100) should be ~0.0: {sigmoid(z_small)}")
    print(f"Sigmoid(0)    should be 0.5:  {sigmoid(z_zero)}")
```

## Sigmoid Function Graph

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 1. Generate range of Z values (from -10 to 10)
z = np.linspace(-10, 10, 100)

# 2. Compute Sigmoid values
y = sigmoid(z)

# 3. Plotting
plt.figure(figsize=(10, 8))
plt.plot(z, y, color='blue', label=r'$\sigma(z) = \frac{1}{1+e^{-z}}$')

# Visual Aids from the Transcript
plt.axhline(0, color='black', linewidth=1) # Lower bound (0)
plt.axhline(1, color='red', linestyle='--', alpha=0.5, label="Max Probability (1.0)")
plt.axhline(0.5, color='green', linestyle=':', label="Decision Boundary (0.5)")
plt.axvline(0, color='black', linewidth=0.5)

plt.title("Sigmoid Activation Function")
plt.xlabel("z (Linear Input)")
plt.ylabel("$\sigma(z)$ (Probability)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
