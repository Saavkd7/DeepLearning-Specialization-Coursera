## 5. Implementation (The "Naive" Way)

Here is the code exactly as described in the transcript. **Note: This is for educational understanding only. Do not use this in production.**



```python
import numpy as np

def naive_gradient_descent_step(w, b, X, Y, learning_rate):
    """
    Computes one step of Gradient Descent using a FOR LOOP (Inefficient).
    """
    m = X.shape[1] # Number of examples
    n_x = X.shape[0] # Number of features
    
    # 1. Initialize Accumulators
    J_sum = 0
    dw = np.zeros((n_x, 1))
    db = 0
    
    # 2. THE BAD LOOP (Iterate over every single example)
    for i in range(m):
        # Extract single example (Shape: n_x, 1)
        x_i = X[:, i].reshape(-1, 1)
        y_i = Y[:, i] # Scalar
        
        # Forward Pass
        z_i = np.dot(w.T, x_i) + b
        a_i = 1 / (1 + np.exp(-z_i)) # Sigmoid
        
        # Accumulate Cost
        J_sum += -(y_i * np.log(a_i) + (1 - y_i) * np.log(1 - a_i))
        
        # Backward Pass (Per example)
        dz_i = a_i - y_i
        
        # Accumulate Gradients
        dw += x_i * dz_i
        db += dz_i

    # 3. Average
    J = J_sum / m
    dw = dw / m
    db = db / m
    
    # 4. Update Parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db
    
    return w, b, J
```
